```python
class MeanEncoder:
    
    def __init__(self, features):
        self.features = features

    def fit(self, X, y):
        self.dct = {}
        X = X.copy()
        X['target'] = y
        for ft in self.features:
            self.dct[ft] = dict(X.groupby(ft)['target'].mean())
        return self

    def transform(self, X):
        X = X.copy()
        for ft in self.features:
            X[ft] = X[ft].map(lambda x: self.dct[ft][x])
        return X
        
    def fit_transform(self, X, y):
        self.fit(X, y)
        return self.transform(X)
```

```python
params = yaml.safe_load(open('../params.yaml'))

TRAIN_PATH = params['path']['train']
TEST_PATH = params['path']['test']
MODEL_PATH = params['path']['model']

models_list = {'Ridge': Ridge(random_state=42),
               'Lasso': Lasso(random_state=42),
               'RandomForest': RandomForestRegressor(random_state=42),
               'GradientBoosting': GradientBoostingRegressor(random_state=42),
               'SVC': SVC(random_state=42)
}

COUNTRIES = params['countries']

max_lag = params['data_params']['max_lag']
test_size = params['data_params']['test_size']
splits_count = params['data_params']['splits_count']
start_split_rate = params['data_params']['start_split_rate']
```

```python
def collect_data(path):
    
    if not os.path.isdir(path):
        raise(Exception('Directory does not exist'))
    
    columns = ['country', 'customer_id', 'invoice', 'price', 'stream_id', 'times_viewed', 'year', 'month', 'day']
    json_list = [i for i in os.listdir(path) if re.fullmatch(r'.+\.json$', i)]

    data = pd.DataFrame(columns=columns)

    if len(json_list) < 1:
        raise(Exception('No file to load'))
       
    
    for file_name in json_list:
        df = pd.read_json(os.path.join(path, file_name))
        df.columns = columns
        data = data.append(df)
    
    data['date'] = pd.to_datetime(data.year.astype('str')+'-'+data.month.astype('str')+'-'+data.day.astype('str'))
    data['country'] = data.country.str.lower()
           
    return data.sort_values(by='date').reset_index(drop=True)
```

```python
def convert_data(original_data, country, mode='train'):
    
    df = original_data.copy()

    start_date = df.date.min()
    end_date = df.date.max()
    dates_range = pd.date_range(start_date, end_date)
    
    if country!='all':
        if country not in df.country.unique():
            raise Exception('Country not found')
        df = df[df.country==country]
    
    df['invoice'] = df.invoice.str.extract(r'([0-9]+)')
    
    
    df_to_join = pd.DataFrame({'date': dates_range})
    df_to_join['month'] = df_to_join.date.dt.strftime('%Y-%m')

    
    
    df = df.groupby('date').agg({'country': len,
                                 'invoice': lambda x: len(np.unique(x)),
                                 'stream_id': lambda x: len(np.unique(x)),
                                 'times_viewed': np.sum,
                                 'price': np.sum}).reset_index()
    
    df.rename(columns={'invoice': 'unique_invoices',
                       'times_viewed': 'total_views',
                       'price': 'revenue',
                       'country': 'purchases',
                       'stream_id': 'total_streams'}, inplace=True)
  
    
    
    df = pd.merge(df_to_join, df, how='left', on='date').fillna(0)
    
#     if mode == 'train':
#         crit_max = df.revenue.mean() + 3*(np.percentile(df.revenue, 75) - np.percentile(df.revenue, 25))
#         df.loc[df.revenue < 0, 'revenue'] = df[df.revenue>=0].revenue.mean()
#         df.loc[df.revenue>crit_max, 'revenue'] = crit_max

    if mode == 'train':
        crit_max = df.revenue.mean() + 3*(np.percentile(df.revenue, 75) - np.percentile(df.revenue, 25))
        t = df[(df.revenue <= crit_max) & (df.revenue >= 0)]
        tst = df[(df.revenue > crit_max) | (df.revenue < 0)]

        if tst.shape[0] > 0:
        
            x_t = t.drop(['revenue', 'date', 'month'], axis=1)
            y_t = t.revenue

            x_tst = tst.drop(['revenue', 'date', 'month'], axis=1)
            y_tst = tst.revenue
        

            model = RandomForestRegressor(random_state=42, max_depth=7, n_estimators=300)
            model.fit(x_t, y_t)

            y_predict = model.predict(x_tst)
        
            df.loc[tst.index, 'revenue'] = y_predict
        
    return df
```

```python
def get_data(path, country=None, renew=False, mode='train'):
    
    data_csv_path = os.path.join(path, 'convert_data')
    
    if not os.path.isdir(data_csv_path):
        os.mkdir(data_csv_path)
           
    countries_to_csv = set(COUNTRIES) if renew \
                    else set(COUNTRIES).difference([i.split('.')[0] for i in os.listdir(data_csv_path)]) 
    
    if len(countries_to_csv) > 0:
        print(f'...Collect json files for {mode}')
        df = collect_data(path)
        print(f'...Create csv files for {mode}')
        for c in countries_to_csv:
            data_country = convert_data(df, c, mode=mode)
            data_country.to_csv(os.path.join(data_csv_path, c)+'.csv', index=False)
            print(f'......csv file for {c} created successfuly')
        print(f'...all csv files for {mode} created successfuly\n')
    
    if country:
        return {country: pd.read_csv(os.path.join(data_csv_path, country.lower())+'.csv', parse_dates=['date'])}
        
    return {i.split('.')[0]: pd.read_csv(os.path.join(data_csv_path, i), parse_dates=[date]) for i in os.listdir(data_csv_path)}
```

```python
def engenier_features(data):
    data['date'] = data.date.astype('datetime64[D]')
    dates = data.date.values.astype('datetime64[D]')
    start = dates.min() - np.timedelta64(max_lag, 'D')
    end = dates.max() + np.timedelta64(1, 'D')
    dates = np.arange(start, end)
    new_data = pd.DataFrame({'date':dates})
    new_data = pd.merge(new_data, data[['date', 'revenue']], how='left', on='date').fillna(0)
        
    features = defaultdict(list)
    features['date'] = dates[:-29]
        
    for date in dates[:-29]:
        cur = date
        end_date = cur + np.timedelta64(30, 'D')
        dates_range = np.arange(cur, end_date)
        mask = np.in1d(dates, dates_range)
        features['y'].append(new_data.loc[mask, 'revenue'].sum())
            
    d = pd.DataFrame(features)    
    d['day_of_week'] = d.date.dt.day_of_week
    d['day'] = d.date.dt.day
    d['month'] = d.date.dt.month
        
    for i in range(1, max_lag+1):
        temp_data = d[['y', 'date']].copy()
        temp_data[f'y_lag_{i}'] = temp_data.y.shift(i)
        temp_data.drop('y', inplace=True, axis=1)
        d = pd.merge(d, temp_data, how='left', on='date')
    d.dropna(inplace=True)
    
    d['month'] = d.month.map(lambda x: 1 if x in [9,10,11] else 0)
    d.set_index('date', inplace=True)
    
    return d.drop('y', axis=1), d.y
```

```python
def get_last_model_version(path):
    return max([int(re.findall(r'\d+', i.split('_')[5])[0]) for i in os.listdir(path)])
```

```python
def set_model_to_test_version(path):
    
    def rename_model(model_name, replace_sample):
        if replace_sample == ('SL__', 'TEST__'):
            print(f'...Set {candidate} as test model')
        old_model_name = os.path.join(path, model_name)
        new_model_name = os.path.join(path, model_name.replace(*replace_sample))
        os.rename(old_model_name, new_model_name)

    
    full_models_list = os.listdir(path)
    sl_models_list = [i for i in full_models_list if i.startswith('SL__')]

    sl_models_errors = [int(re.findall(r'\d+', i.split('_')[6])[0]) for i in sl_models_list]
    candidate = sl_models_list[np.argmin(sl_models_errors)]
    
    
    if len([i for i in full_models_list if i.startswith('TEST__')]) == 0:
        rename_model(candidate, ('SL__', 'TEST__'))
    else:
        current_test_model = [i for i in full_models_list if i.startswith('TEST__')][0]
        current_test_model_error = int(re.findall(r'\d+', current_test_model.split('_')[6])[0])
        candidate_error = np.min(sl_models_errors)    
        
        if candidate_error < current_test_model_error:
            rename_model(candidate, ('SL__', 'TEST__'))
            rename_model(current_test_model, ('TEST__', 'SL__'))    
        else: 
            print('...The quality of the new models is worse than the current test model')
```

```python
def save_model(model, start_name, path, error):
       
    if not os.path.isdir(path):
        os.mkdir(path)
        model_version = 0
    else:
        if len(os.listdir(path)) == 0:
            model_version = 0
        else:
            model_version = get_last_model_version(path)
    
    date_tag = datetime.now().strftime('%Y%m%d_%H%M%S')
    
    model_name = f'SL__{start_name}_{date_tag}_v{model_version+1}_err{int(error)}.joblib'
    
    dump(model, os.path.join(path, model_name))
```

```python
def load_model(country, version=None):
    
    path = os.path.join(MODEL_PATH, country.lower())
    
    if country.lower() not in COUNTRIES:
        print(f'...For this country no models. You can load models for countries in this list:')
        print('......', ', '.join(COUNTRIES))
        return None
    
    if not os.path.isdir(path):
        os.mkdir(path)
    
    test_list = [i for i in os.listdir(path) if i.startswith('TEST__')]   
    
    if version:
        models_vers = [re.findall(r'\d+', i.split('_')[5])[0] for i in os.listdir(path)]
        if str(version) not in models_vers:
            print(f'...Model with version {version} is not exist')
            return None
        model_name = os.listdir(path)[models_vers.index(str(version))]

    else:
        if len(test_list) == 0:
            fit_models(country)
        test_list = [i for i in os.listdir(path) if i.startswith('TEST__')]
        model_name = test_list[0]
        
    model = load(os.path.join(path, model_name))
    return model
```

```python
def fit_models_(data, path, test=False):
    
    train_val_split_index = int(d.shape[0]*(1-test_size))

    X, y = engenier_features(data)
    
    X_train, y_train = X[:train_val_split_index], y[:train_val_split_index]
    X_val, y_val = X[train_val_split_index:], y[train_val_split_index:]

    cv_split_index = np.linspace(int(start_split_rate*X_train.shape[0]), 
                                 X_train.shape[0], 
                                 splits_count+1).astype('int')


    cv = [[np.arange(cv_split_index[i]), np.arange(cv_split_index[i], cv_split_index[i+1])]
                                  for i in range(len(cv_split_index)-1)]
    
    
    for model_name in params['models']:
        print(f'......Train {model_name} model')
        model = models_list[model_name]
       
        pipline =Pipeline(steps=[('MeanEncoder', MeanEncoder(features=['day_of_week', 'day'])),
                                 ('scaler', StandardScaler()),
                                 (model_name, model)])
        
        param_grid = params['models'][model_name]
        
        gcv = GridSearchCV(estimator=pipline,
                           param_grid=param_grid,
                           cv=cv, 
                           scoring='neg_root_mean_squared_error')
        
        gcv.fit(X_train, y_train)
        best_estimator = gcv.best_estimator_
       
        preds = best_estimator.predict(X_val)
        error = np.sqrt(mean_squared_error(preds, y_val))
        
        print(f'.........RMSE for {model_name} on validation = {error}')
        
        best_estimator.fit(X, y)
        
        save_model(best_estimator, model_name, path, error)
        
    set_model_to_test_version(path)
```

```python
# Функция, которая обучвет модель и слхраняет ее в папке со страной

def fit_models(country=None, renew_data=False):
    
    country = country.lower() if country else country
    dfs = get_data(TRAIN_PATH, country, renew_data)
    
    for country, df in dfs.items():
        print(f'...Train models for country: {country}')
        path = os.path.join(MODEL_PATH, country)
        fit_models_(df, path)
        print(f'...Models trained successfuly\n')
```

```python
def predict(country, day, month, year, model_version=None):

    model = load_model(country, version=model_version)
    
    data = get_data(TRAIN_PATH, country=country)[country]
    y = engenier_features(data)[1]
   
    target_date = pd.Timestamp(year=year, month=month, day=day)
    start_date_to_predict = y.index.max() + timedelta(1)
    dates_range_to_predict = pd.date_range(start_date_to_predict, target_date)
    
    for date_to_predict in dates_range_to_predict:
        
        x = np.hstack((np.array([date_to_predict.day_of_week, 
                                 date_to_predict.day, 
                                 1 if date_to_predict.month in [9,10,11] else 0]), y[-max_lag:][::-1]))
        x = pd.DataFrame([x], 
                         columns=['day_of_week', 'day', 'month'] + [f'y_lag_{i}' for i in range(1, max_lag+1)])
        
        y_predicted = model.predict(x)
        print(y_predicted)
        y = y.append(pd.Series(y_predicted, index=[date_to_predict]))
    
    
    return y[start_date_to_predict:]
```
